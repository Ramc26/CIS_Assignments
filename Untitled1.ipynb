```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
site="https://www.rrc.texas.gov/resource-center/research/data-sets-available-for-download/"
page=requests.get(site)
```

```python
print(page.text)
```

```python
soup=bs(page.content,"html.parser")
soup
```

```python
results=soup.find('table',class_="resTab table-hover")
print(results)
```
